```python
import hashlib
def hash_sha256(data, upper: bool = False) -> str:
    if (type(data) == type('')):
        data = data.encode('utf-8')
    if upper:
        return hashlib.sha256(data).hexdigest().upper()
    return hashlib.sha256(data).hexdigest()

print(hash_sha256(head))
print(hash_sha256(tail))
```
